## Insertion 
### Check tree is empty or not? Yes or NO
**if yes then node will insert in the first node of the tree**
**if no means tree contains already some nodes so we need to specify in which place to insert new node.**
**Position of new_node Either we need to place into left subtree or right subtree.**

```python
class BST:
    """
    This __init__ will initialize the object automatically whenever you create an object.
    """
    def __init__(self, key): 
        self.key = key
        self.lchild = None
        self.rchild = None
    
    def insert(self, data):
        
        """
        If root.key is empty then add root.key = data
        """
        if self.key is None:
            self.key = data
            return
        """
        If root.key is equalsto data then return
        """
        if self.key == data:
            return
        """
        If root.key is greater than data move to left subtree,
        """
        if self.key > data:
            """
            If here lchild is already present then add or on the basis of the value 
            So that we need to use here recursive function
            and then save the value in it
            """
            if self.lchild:
                self.lchild.insert(data)
            else:
                """
                If not lchild present this data will become as the lchild
                """
                self.lchild = BST(data)
                """
                If root.key is less than the data so it will go to the right side of the root.key
                """
        else:
            """
            If here rchild is already present then add or on the basis of the value 
            So that we need to use here recursive function
            and then save the value in it
            """
            if self.rchild:
                self.rchild.insert(data)
            
            else:
                """
                If not lchild present this data will become as the lchild
                """
                self.rchild = BST(data)

"""
Object creating for the Binary search tree.
"""
## Here we created and the output is like lchild -> None, key -> 10, rchild -> None
bst1 = BST(None)
bst1.insert(10)
```

```python
class BST:
    """
    This __init__ will initialize the object automatically whenever you create an object.
    """
    def __init__(self, key): 
        self.key = key
        self.lchild = None
        self.rchild = None
    
    def insert(self, data):
        
        """
        If root.key is empty then add root.key = data
        """
        if self.key is None:
            self.key = data
            return
        """
        If root.key is equalsto data then return
        """
        if self.key == data:
            return
        """
        If root.key is greater than data move to left subtree,
        """
        if self.key > data:
            """
            If here lchild is already present then add or on the basis of the value 
            So that we need to use here recursive function
            and then save the value in it
            """
            if self.lchild:
                self.lchild.insert(data)
            else:
                """
                If not lchild present this data will become as the lchild
                """
                self.lchild = BST(data)
                """
                If root.key is less than the data so it will go to the right side of the root.key
                """
        else:
            """
            If here rchild is already present then add or on the basis of the value 
            So that we need to use here recursive function
            and then save the value in it
            """
            if self.rchild:
                self.rchild.insert(data)
            
            else:
                """
                If not lchild present this data will become as the lchild
                """
                self.rchild = BST(data)
                
                
    def search(self, data):
                
        """
        If root.key is empty then add root.key = data
        """
        if self.key == data:
            print("Node is Found")
            return
        
        if data < self.key:
            """
            If here lchild is already present then add or on the basis of the value 
            So that we need to use here recursive function
            and then save the value in it
            """
            if self.lchild:
                self.lchild.search(data)
                print("left side")
            else:
                """
                If not then Node is not present in tree
                """
                print("Node is not present in tree!!")

        else:
            """
            If root.key is less than the data so it will go to the right side of the root.key
            
            If here rchild is already present then add or on the basis of the value 
            So that we need to use here recursive function
            and then save the value in it
            """
            if self.rchild:
                self.rchild.search(data)
                print("right side")
                
            else:
                """
                If not lchild present this data will become as the lchild
                """
                print("Node is not present in tree!!")
                
    
    
    def pre_order(self):
        """
        `root`, `left`, `right`
        """
#         print(self.key)
        print(self.key, end = " ")
        ## Left child is present mean use the recursion to find the order of left sub tree
        if self.lchild:
            self.lchild.pre_order()
            
        ## Right child is present mean use the recursion to find the order of right sub tree
        if self.rchild:
            self.rchild.pre_order()
    
    def in_order(self):
        """
        `left`, `root`, `right`
        """
        
        ## Left child is present mean use the recursion to find the order of left sub tree
        if self.lchild:
            self.lchild.in_order()
            
#         print(self.key)
        print(self.key, end = " ")
            
        ## Right child is present mean use the recursion to find the order of right sub tree
        if self.rchild:
            self.rchild.in_order()

    def post_order(self):
        """
        `left`, `right`, `root`
        """
        
        ## Left child is present mean use the recursion to find the order of left sub tree
        if self.lchild:
            self.lchild.post_order()
            
            
        ## Right child is present mean use the recursion to find the order of right sub tree
        if self.rchild:
            self.rchild.post_order()
            
#         print(self.key)
        print(self.key, end = " ")
    
    
    def delete(self, data, curr):
#     def delete(self, data):
        
        if self.key is None:
            print("Tree is Empty")
            return
        if data < self.key:
            if self.lchild:
                self.lchild = self.lchild.delete(data, curr)
#                 self.lchild = self.lchild.delete(data)
            else:
                print("Givben node is not present in the tree")
        elif data > self.key:
            if self.rchild:
                self.rchild = self.rchild.delete(data, curr)
#                 self.rchild = self.rchild.delete(data)
                
            else:
                print("Given node is not present in the tree!!")
        else:
            if self.lchild is None:
                temp = self.rchild
                if data == curr:
                    self.key = temp.key
                    self.lchild = temp.lchild
                    self.rchild = temp.rchild
                    temp = None
                    return
            if self.rchild is None:
                temp = self.lchild
                if data == curr:
                    self.key = temp.key
                    self.lchild = temp.lchild
                    self.rchild = temp.rchild
                    temp = None
                    return
            node = self.rchild
            while node.lchild:
                node = node.lchild
            self.key = node.key
            self.rchild = self.rchild.delete(node.key, curr)
#             self.rchild = self.rchild.delete(node.key)
            
        return self
    
"""
Object creating for the Binary search tree.
"""
## Here we created and the output is like lchild -> None, key -> 10, rchild -> None
root = BST(10)
lst = [6, 3, 1, 3, 98, 8, 7]
for i in lst:
    root.insert(i)
# root.search(10)
# root.search(100)
# root.search(98)
# root.search(7)

# root.pre_order()
## `root`, `left`, `right`

# root.in_order()
# # `left`,`root`, `right`

# print("post_order")
# root.post_order()
# `left` , `right` ,`root`


# root.delete(3, 6)
```
